```python
# Descarga de boletin epidemiologico
# @elvagodeldato

# Librerias
import os
import re
import PyPDF2
import pytesseract
import cv2

from PyPDF2 import PdfFileReader, PdfFileWriter
from PIL import Image
from pdf2image import convert_from_path
```

```python
def archivos_pdf(filefolder):
    archivos = []
    for i in os.scandir(filefolder):
        #print(i)
        a = str(i).replace("<DirEntry","/")
        b = str(a).replace("'","")
        c = b.replace(">","")
        d = c.replace(" ","")
        e = str(filefolder) + d
        archivos.append(e)
    return archivos
```

```python
def get_pag():
    for foldername,subfolders,files in os.walk(bases):
        for file in files:
            #file = PdfFileReader(p8)
            pdf = PdfFileReader(os.path.join(foldername,file))
            doc = PdfFileWriter()
            doc.addPage(pdf.getPage(7))
            split = open(os.path.join(foldername+'/pag_8'+ file),'wb') 
            doc.write(split)
            split.close()
    return 'La pagina ha sido creada'
```
